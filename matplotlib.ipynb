```python
x = np.linspace(-3, 3, 50)
y1 = 2*x +1
y2 = x**2
plt.figure(num = 3, figsize=(4, 3))
plt.plot(x, y1)
plt.figure(figsize=(4, 4))
plt.plot(x, y2)
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.plot(x, y2)
plt.plot(x, y1, color = 'Red', linewidth = 2, linestyle ='--')
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.plot(x, y2)
plt.plot(x, y1, color = 'Red', linewidth = 2, linestyle ='--')
plt.xlim((-1,2))
plt.ylim((-1,4))
plt.xlabel('class numbers')
plt.ylabel('marks numbers')
plt.xticks(np.linspace(-1,2,4))
plt.yticks([-1,1,3,4],['$wrose$','bad','good','well'])
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.xlim((-1,2))
plt.ylim((-1,4))
plt.xlabel('class numbers')
plt.ylabel('marks numbers')
plt.xticks(np.linspace(-1,2,4))
plt.yticks([-1,1,3,4],['$wrose$','bad','good','well'])
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data','0'))
l1,= plt.plot(x, y2, label = 'up')
l2,= plt.plot(x, y1, color = 'Red', linewidth = 2, linestyle ='--',label='down ')
plt.legend(handles=[l1,l2,],loc='best')
plt.show()
```

```python
plt.figure(num=0,figsize=(8,5))
x = np.linspace(-3,3,50)
y = 2*x + 1
plt.xlim((-3,3))
plt.ylim((-6,8))
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.linspace(-3,3,7))
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
l,=plt.plot(x,y,color='red',label='text')
plt.legend(handles=[l,],loc='best')
x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0,y0,color ='blue')
plt.plot([x0,x0],[y0,0],'k--',lw=2.5)
plt.annotate(r'$2x+1=%s$'%y0,xy=(x0,y0),xycoords='data',xytext=(+30,-30),textcoords='offset points',fontsize=16,arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))
plt.text(-3,3,r'$this\ is\ a\ good\ text.\mu\sigma_i\alpha_t$',fontsize=16,color='purple')
plt.show()
```

```python

n = 284
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))
plt.scatter(X,Y,s=125,alpha=0.5,c=T)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
n = 12
x = np.arange(12)
y1 = (1-x/float(n))*np.random.uniform(0.5,1.0,n)
y2 = (1-x/float(n))*np.random.uniform(0.5,1.0,n)
plt.bar(x,+y1,facecolor = 'red',edgecolor = 'white',alpha = 0.7)
plt.bar(x,-y2,facecolor = 'blue',edgecolor = 'black')
plt.xticks(())
plt.yticks(())
for x1,y in zip(x,y1):
    plt.text(x1-0.5,y+0.05,'%.2f'%y,color='blue')
for x2,y in zip(x,y2):
    plt.text(x2-0.5,-y-0.07,'-%.2f'%y,color='red')
plt.show()
```

```python
def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
plt.contourf(X,Y,f(X,Y),8,alpha = 0.75,cmap = plt.cm.hot)
c = plt.contour(X,Y,f(X,Y),8,colors='black',linewidths=1.5)
plt.clabel(c,inline=False,fontsize = 10)

plt.xticks(())
plt.yticks(())
plt.show()
```
